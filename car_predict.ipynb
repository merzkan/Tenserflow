verilerimizi temizledik/// model oluşturma 

y ulaşmak istedigimiz şey

```python
y = df["price"].values # numpy dizisine çevirdik
x = df.drop("price",axis=1).values
```

x ve y dizilerimizi eğiticegimiz ve test edecegimiz diziye bölüyoruz

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

scaling işlemimizi yapacagız

```python
from sklearn.preprocessing import  MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

modeli oluşturmak

```python
from tensorflow.keras.models import Sequential #katmanları oluşturuyoruz
from tensorflow.keras.layers import Dense #layersleri oluşturuyoruz
```

```python
x_train.shape
```

```python
model = Sequential()
model.add(Dense(12,activation="relu")) #12nöron
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
```

modeli eğitme training

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

Epoch: Tüm veri kümesinin modelden bir kez geçmesi.


Batch: Bir epoch'un içindeki alt gruplardan biri. Eğer 100 batch varsa, bir epoch'ta model 100 batch üzerinde işlem yapar.Batch, makine öğrenimi ve derin öğrenmede, eğitim veri kümesini daha küçük alt gruplara ayırmak için kullanılan bir terimdir. Bu, modelin eğitim sürecinde daha verimli çalışmasını ve bellek yönetiminin optimize edilmesini sağlar.

Örnekle Açıklama:
Bir veri kümesi düşünelim:

Toplam 10.000 veri örneğiniz var.
Modeli eğitirken, bu verilerin tümünü aynı anda işlemek yerine, onları küçük gruplara (örneğin 100'erlik gruplar) bölebilirsiniz.
Bu küçük gruplara batch denir. Eğer her bir batch 100 veri içeriyorsa:

Batch size (batch boyutu) = 100 olur.
Toplam batch sayısı = 10.000 veri / 100 batch = 100 batch.
Neden Batch Kullanılır?
Bellek Yönetimi: Tüm veri kümesini bir anda işlemeye çalışmak, özellikle büyük veri kümelerinde, bilgisayarın belleğini aşabilir. Batch kullanarak bu sorun çözülür.
Eğitim Verimliliği: Her bir batch üzerinde modelin ağırlıklarını güncellemek, tüm veri kümesini işlemekten daha hızlıdır.
Stokastik Optimizasyon: Küçük batch'lerle eğitim, modelin daha iyi genelleme yapmasına yardımcı olabilir çünkü her batch biraz farklı veri örnekleri içerir.
Batch Size Seçimi
Küçük Batch Size (örneğin, 16, 32): Daha iyi genelleme yapabilir ama eğitim daha yavaş olabilir.
Büyük Batch Size (örneğin, 128, 256): Daha hızlı eğitim sağlar ama genelleme performansı biraz düşebilir.

model.fit():

Modelin eğitimini başlatan fonksiyon.
Verilen eğitim verilerini kullanarak ağırlıkları günceller ve öğrenme işlemini gerçekleştirir.

Parametreler:
x=x_train, y=y_train
x_train: Modelin giriş (bağımsız değişkenler) verileri.
y_train: Modelin çıkış (etiket veya hedef) verileri.
Bu ikili, eğitim için kullanılan veri setidir.

validation_data=(x_test, y_test)
x_test ve y_test: Test veri seti (giriş ve hedef verileri).
Eğitim sırasında, model bu verileri kullanarak doğrulama (validation) yapar.
Doğrulama, modelin her epoch sonunda bu test verileri üzerinde nasıl bir performans sergilediğini ölçmek için kullanılır. Eğitim sırasında model, doğrulama verisiyle optimizasyon yapmaz.

batch_size=250
Her iterasyonda kaç veri örneği üzerinde birden modelin ağırlıklarını güncelleyeceğini belirler.
Burada batch size = 250 olduğu için:
Eğitim veri seti 250'lik gruplara ayrılır.
Her batch işlendiğinde model, bu batch üzerinden hata hesaplar ve ağırlık güncellemesi yapar.
Örneğin, eğitim veri seti 10,000 örnek içeriyorsa, her epoch'ta 
10,000/250=40 =40 batch olacak.

epochs=300
Tüm eğitim veri setinin model tarafından kaç kez işleneceğini belirler.
300 epoch demek, modelin tüm eğitim verisini 300 kez işlemesi anlamına gelir
