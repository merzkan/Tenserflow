```python
y = df["Type"].values
x = df.drop("Type",axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

Earlystopping ve Dropout yöntemlerini overfitting proplemerini çözmek için

Overfitting (Aşırı Uydurma): bir makine öğrenimi modelinin eğitim verisine çok iyi uyum sağladığı, ancak yeni, görülmemiş veriler üzerinde düşük performans gösterdiği bir durumdur. 

```python
x_train.shape
```

```python
model = Sequential()
model.add(Dense(units=30,activation ="relu"))#units içeri 30tane nöron koy
model.add(Dense(units=15,activation ="relu"))
model.add(Dense(units=15,activation ="relu"))
model.add(Dense(units=1,activation ="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

binary_crossentropy, derin öğrenme modellerinde kullanılan bir kayıp fonksiyonudur. Bu kayıp fonksiyonu, özellikle ikili sınıflandırma (binary classification) problemleri için uygundur. Modelin, her veri örneğinin bir sınıfa (örneğin, 0 veya 1) ait olma olasılığını tahmin etmesine yardımcı olur.

Çalışma Mantığı
Tahmin edilen çıktı (y): Modelin 0 ile 1 arasında bir olasılık değeri döndürmesi beklenir (örneğin, bir sigmoid aktivasyon fonksiyonu kullanılırsa).

Gerçek çıktı (y): Etiketler 0 veya 1 olarak kodlanır.

Kayıp fonksiyonu, modelin doğru sınıf için yüksek bir olasılık tahmin etmesini teşvik eder.

Amaç
Modelin tahmin ettiği olasılıklar ile gerçek etiketler arasındaki farklılığı minimize etmek.

Neden Kullanılır?
İkili sınıflandırmada modelin yanlış tahminlerini daha fazla cezalandırarak doğru sınıfa daha yakın tahmin yapmasını sağlar.

1.Sınıflandırma Problemleri İçin:
binary_crossentropy:
İkili sınıflandırma problemleri için.

2.Regresyon Problemleri İçin:

mse (Mean Squared Error - Ortalama Kare Hata):
Sürekli değer tahmini için.
Hata karelerinin ortalamasını hesaplar

mae (Mean Absolute Error - Ortalama Mutlak Hata):
Sürekli değerler için kullanılır, tahmin ile gerçek değer arasındaki mutlak farkın ortalamasını alır.

val_loss saçma sapan artarsa o zaman epochu durdur o zaman traning etmene gerek yok EarlyStopping o zaman işe yarar.

Early Stopping(erken durdurma) adı verilen bir eğitim süreci kontrol mekanizmasını tanımlar. 
kerasta eğitim sırasında modelin performansını izlemek ve gereksiz
yere fazla epoch çalışmasını engellemek için kullanılır. 

```python
model = Sequential()
model.add(Dense(units=30,activation ="relu"))#units içeri 30tane nöron koy
model.add(Dense(units=15,activation ="relu"))
model.add(Dense(units=15,activation ="relu"))
model.add(Dense(units=1,activation ="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping = EarlyStopping(monitor ="val_loss",mode="min",verbose=1,patience=25)
```

1.EarlyStopping
Keras'taki bir eğitim callback (geri çağırım) sınıfıdır.
Eğitim sırasında bir metrik izlenir ve belirli bir koşul sağlandığında eğitim otomatik olarak durdurulur.

2.monitor="val_loss"
val_loss: Doğrulama kaybını izler. Eğitim sırasında her epoch'ta doğrulama verilerindeki kayıp (loss) değerine bakar.

3.mode="min"
İzlenen metrik için hedef yönü belirler.
"min": İzlenen değerin mümkün olduğunca azalması istenir (örneğin, kayıp - loss - değerleri).
"max": İzlenen değerin mümkün olduğunca artması istenir (örneğin, doğruluk - accuracy).
"auto": Keras, kullanılan metrik türüne göre otomatik karar verir (val_loss için genellikle "min" seçilir).

4.patience=25
Bekleme süresi.
Eğer izlenen metrik (örneğin val_loss), 25 epoch boyunca iyileşme göstermezse eğitim durdurulur.

```python
model.fit(x=x_train,y=y_train,epochs = 700,validation_data = (x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

Dropout, derin öğrenme modellerinde kullanılan bir düzenleme (regularization) tekniğidir. 
Amacı modelin aşırı öğrenmesini (overfitting) engellemektir.
Dropout, eğitim sırasında belirli nöronları rastgele devre dışı bırakarak çalışır.

```python
model = Sequential()
model.add(Dense(units=30,activation ="relu"))#units içeri 30tane nöron koy
model.add(Dropout(0.50))
model.add(Dense(units=15,activation ="relu"))
model.add(Dropout(0.50))
model.add(Dense(units=15,activation ="relu"))
model.add(Dropout(0.50))

model.add(Dense(units=1,activation ="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs = 700,validation_data = (x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
```
