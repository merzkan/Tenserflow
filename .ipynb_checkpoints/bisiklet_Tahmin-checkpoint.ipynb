veriyi test/train olarak ikiye ayÄ±rmak

```python
from sklearn.model_selection import train_test_split
```

```python
#y = wx + b
#y -> label
y = dataFrame["Fiyat"].values ## diziye Ã§eviriyoruz

#x ->feature(Ã¶zellik)
x = dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values # values dersem numpy dizisi olur
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=15)
```

```python
x_train.shape#670tane veri buraya geldi
```

```python
x_test.shape #330tane veri buraya geldi
```

```python
y_train.shape
```

```python
y_test.shape
```

#scaling 0ve1 arasÄ±nda sayÄ±lara Ã§eviriliyor aynÄ± oranlarda( Ã¶zellikle x verilerimizi)Veri Ã¶lÃ§eklendirme, makine Ã¶ÄŸrenimi algoritmalarÄ±nÄ±n daha verimli ve doÄŸru Ã§alÄ±ÅŸabilmesi iÃ§in verinin belirli bir aralÄ±kta normalize edilmesi iÅŸlemidir.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler  = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
#x_test: Modelin test verisi (Ã¶zellikler).
#y_test: Test verisinin doÄŸru sonuÃ§larÄ± (etiketler).
#x_train: Modelin giriÅŸ verileri (Ã¶zellikler, yani baÄŸÄ±msÄ±z deÄŸiÅŸkenler).
#y_train: Modelin doÄŸru sonuÃ§larÄ± (etiketler, yani baÄŸÄ±mlÄ± deÄŸiÅŸkenler).
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model =Sequential()#modelimizi bu sÄ±nÄ±ftan oluÅŸturuyoruz. ondan sonra hangi katmanlarla Ã§alÄ±ÅŸcaz onu belirtiriyoruz.

model.add(Dense(4,activation="relu"))#hidden layer
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))

model.add(Dense(1))#output katmanÄ±

model.compile(optimizer="rmsprop",loss ="mse")
```

Model oluÅŸturduk traning baÅŸlÄ±caz

```python
model.fit(x_train,y_train,epochs=250) #250kere tekrar edecek
```

```python
loss =model.history.history["loss"]
```

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

```python
trainLoss = model.evaluate(x_train,y_train)
```

 Bu fonksiyonlar, modelinizin Ã¶ÄŸrenme sÃ¼reci sÄ±rasÄ±nda ne kadar iyi genel performans gÃ¶sterdiÄŸini anlamanÄ±zÄ± saÄŸlar.
Genellikle, eÄŸitim kaybÄ± zamanla azalÄ±rken, test kaybÄ± da azalmaya devam etmelidir. EÄŸer test kaybÄ± artarsa, modelin aÅŸÄ±rÄ± uyum yapÄ±yor olabileceÄŸini gÃ¶sterir.

```python
testLoss = model.evaluate(x_test,y_test)
```

```python
mean_absolute_error(tahminDf["GerÃ§ek Y"],tahminDf["Tahmin Y"])#hata payÄ±  #Mean Absolute Error (MAE) - Ortalama Mutlak Hata:
```

MAE: GerÃ§ek deÄŸerler ile tahmin edilen deÄŸerler arasÄ±ndaki farklarÄ±n mutlak deÄŸerlerinin ortalamasÄ±nÄ± alÄ±r. Yani her bir hata pozitif yapÄ±lÄ±r, yÃ¶nÃ¼ Ã¶nemli deÄŸildir.
KÃ¼Ã§Ã¼k hatalarÄ±n daha Ã¶nemli olduÄŸu durumlarda kullanÄ±lÄ±r.
Modelin genel doÄŸruluÄŸunu daha anlaÅŸÄ±lÄ±r bir ÅŸekilde gÃ¶rmek isterseniz tercih edilebilir.

```python
mean_squared_error(tahminDf["GerÃ§ek Y"],tahminDf["Tahmin Y"])#hata payÄ± #. Mean Squared Error (MSE) - Ortalama Kare Hata:
```

MSE: GerÃ§ek deÄŸerler ile tahmin edilen deÄŸerler arasÄ±ndaki farklarÄ±n karelerinin ortalamasÄ±nÄ± alÄ±r. Bu, bÃ¼yÃ¼k hatalarÄ±n daha fazla cezalandÄ±rÄ±lmasÄ±na neden olur Ã§Ã¼nkÃ¼ farklar karesel olarak bÃ¼yÃ¼r.BÃ¼yÃ¼k hatalarÄ±n daha Ã¶nemli olduÄŸu ve bunlara daha fazla dikkat edilmesi gereken durumlarda kullanÄ±lÄ±r.
MSE'nin daha duyarlÄ± olmasÄ± nedeniyle, modelin bÃ¼yÃ¼k hatalarÄ± dÃ¼zelterek daha iyi genelleme yapmasÄ±nÄ± saÄŸlayabilirsiniz.
